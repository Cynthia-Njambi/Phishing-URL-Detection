```python
import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed
from urllib.parse import urlparse
import warnings
from requests.packages.urllib3.exceptions import InsecureRequestWarning

# Suppress InsecureRequestWarning
warnings.simplefilter('ignore', InsecureRequestWarning)

# Function to check if a website is live, without SSL verification
def is_website_live(url, timeout):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    # Ensure URL has http/https
    if not url.startswith(('http://', 'https://')):
        url = 'http://' + url

    # Try without "www" first
    try:
        response = requests.get(url, headers=headers, timeout=timeout, verify=False)
        if response.status_code == 200:
            return True
    except requests.RequestException:
        pass

    # Retry with "www" if the first attempt fails
    if not urlparse(url).netloc.startswith("www."):
        url_with_www = url.replace("://", "://www.")
        try:
            response = requests.get(url_with_www, headers=headers, timeout=timeout, verify=False)
            return response.status_code == 200
        except requests.RequestException:
            return False

    return False

# Function to process URLs concurrently
def check_urls_concurrently(urls, timeout=10, max_workers=10):
    results = {}
    error_count = 0
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_url = {executor.submit(is_website_live, url, timeout): url for url in urls}
        for future in as_completed(future_to_url):
            url = future_to_url[future]
            try:
                results[url] = future.result()
            except Exception:
                results[url] = False
                error_count += 1

    if error_count > 0:
        print(f"Encountered {error_count} errors in this batch.")

    return results

# Example usage with a batch process
def process_in_batches(data, batch_size=10000, timeout=10, max_workers=100):
    all_results = []
    for i in range(0, len(data), batch_size):
        batch_data = data.iloc[i:i+batch_size]
        batch_urls = batch_data['url'].tolist()
        print(f"Processing batch {i // batch_size + 1} of {len(data) // batch_size + 1}")

        # Get live status results
        batch_results = check_urls_concurrently(batch_urls, timeout=timeout, max_workers=max_workers)

        # Merge results with original batch data
        batch_data = batch_data.copy()  # Avoid modifying original DataFrame
        batch_data['is_website_live'] = batch_data['url'].map(batch_results)

        # Save batch to CSV
        batch_data.to_csv(f'batch_results_{i // batch_size + 1}.csv', index=False)
        all_results.append(batch_data)

    # Combine all results
    final_df = pd.concat(all_results).reset_index(drop=True)
    return final_df

# Assuming your DataFrame is named 'data'
data_sample = data.sample(frac=1, random_state=42).reset_index(drop=True)[:10000]

# Run the process with custom parameters
batch_size = 10000
timeout = 10
max_workers = 10000  # Adjust based on your system's capacity

final_result_df = process_in_batches(data_sample, batch_size=batch_size, timeout=timeout, max_workers=max_workers)
final_result_df.to_csv("all_url_check_results.csv", index=False)
```

```python
import aiohttp
import asyncio
import pandas as pd
from bs4 import BeautifulSoup
import logging
import random
from typing import List, Tuple, Optional
from dataclasses import dataclass
from urllib.parse import urlparse
import time

# Setup logging with more detailed format
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

@dataclass
class ScraperConfig:
    """Configuration parameters for the scraper"""
    retries: int = 1
    timeout: int = 30
    max_concurrent_requests: int = 500
    chunk_size: int = 500
    backoff_min: float = 1.0
    backoff_max: float = 3.0

class URLMetadataFetcher:
    def __init__(self, config: ScraperConfig = ScraperConfig()):
        self.config = config
        self.session: Optional[aiohttp.ClientSession] = None
        self._rate_limits = {}
        self._domain_semaphores = {}

    async def __aenter__(self):
        self.session = aiohttp.ClientSession()
        return self

    async def __aexit__(self, exc_type, exc_val, exc_tb):
        if self.session:
            await self.session.close()

    def _normalize_url(self, url: str) -> str:
        """Ensure URL has proper scheme"""
        if not url.startswith(('http://', 'https://')):
            return 'http://' + url
        return url

    def _get_domain_semaphore(self, url: str) -> asyncio.Semaphore:
        """Get or create a semaphore for a specific domain"""
        domain = urlparse(url).netloc
        if domain not in self._domain_semaphores:
            # Limit to 5 concurrent requests per domain
            self._domain_semaphores[domain] = asyncio.Semaphore(5)
        return self._domain_semaphores[domain]

    async def fetch_metadata(self, url: str) -> Tuple[str, str]:
        """Fetch title and description metadata from a URL with retries and backoff"""
        normalized_url = self._normalize_url(url)
        domain_semaphore = self._get_domain_semaphore(normalized_url)

        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }

        async with domain_semaphore:
            for attempt in range(self.config.retries):
                try:
                    async with self.session.get(
                        normalized_url,
                        headers=headers,
                        timeout=self.config.timeout
                    ) as response:
                        if response.status == 200:
                            html = await response.text()
                            soup = BeautifulSoup(html, 'html.parser')

                            # Extract title
                            title = (soup.title.string.strip() if soup.title
                                   else soup.find('h1').text.strip() if soup.find('h1')
                                   else 'No Title')

                            # Extract description with fallbacks
                            meta_desc = (
                                soup.find('meta', attrs={'name': 'description'}) or
                                soup.find('meta', attrs={'property': 'og:description'}) or
                                soup.find('meta', attrs={'name': 'twitter:description'})
                            )
                            description = meta_desc['content'].strip() if meta_desc else 'No Description'

                            return title, description

                        elif response.status == 429:  # Too Many Requests
                            retry_after = int(response.headers.get('Retry-After', 60))
                            logging.warning(f"Rate limited for {url}. Waiting {retry_after}s")
                            await asyncio.sleep(retry_after)
                            continue

                        else:
                            logging.error(f"Failed to fetch {url}: Status {response.status}")

                except Exception as e:
                    logging.error(f"Error fetching {url} (Attempt {attempt + 1}): {str(e)}")
                    if attempt < self.config.retries - 1:
                        backoff = random.uniform(
                            self.config.backoff_min * (attempt + 1),
                            self.config.backoff_max * (attempt + 1)
                        )
                        await asyncio.sleep(backoff)
                        continue

            return 'No Title', 'No Description'

    async def process_urls(self, urls: List[str]) -> List[Tuple[str, str]]:
        """Process a batch of URLs with concurrency control"""
        semaphore = asyncio.Semaphore(self.config.max_concurrent_requests)

        async def sem_fetch(url: str) -> Tuple[str, str]:
            async with semaphore:
                return await self.fetch_metadata(url)

        return await asyncio.gather(*[sem_fetch(url) for url in urls])

async def process_dataframe(df: pd.DataFrame, url_column: str) -> pd.DataFrame:
    """Process a DataFrame containing URLs and add metadata columns"""
    start_time = time.time()
    urls = df[url_column].tolist()
    config = ScraperConfig()
    results = []

    async with URLMetadataFetcher(config) as fetcher:
        for i, chunk in enumerate(range(0, len(urls), config.chunk_size)):
            chunk_urls = urls[chunk:chunk + config.chunk_size]
            chunk_results = await fetcher.process_urls(chunk_urls)
            results.extend(chunk_results)

            # Progress update
            processed = len(results)
            success_count = sum(1 for r in results if r[0] != 'No Title')
            elapsed = time.time() - start_time
            rate = processed / elapsed if elapsed > 0 else 0

            logging.info(
                f"Processed {processed}/{len(urls)} URLs "
                f"({success_count/processed*100:.1f}% success) "
                f"at {rate:.1f} URLs/sec"
            )

    # Create result DataFrame
    metadata_df = pd.DataFrame(results, columns=['title', 'description'])
    final_df = pd.concat([df.reset_index(drop=True), metadata_df], axis=1)

    # Log final statistics
    success_rate = sum(1 for r in results if r[0] != 'No Title') / len(results) * 100
    logging.info(f"Final Success Rate: {success_rate:.1f}%")
    logging.info(f"Total Time: {time.time() - start_time:.1f} seconds")

    return final_df
```

```python
import asyncio
import pandas as pd
import logging
from typing import List, Tuple, Optional
from whois import whois
from urllib.parse import urlparse
import time

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(message)s')

async def fetch_whois_data(domain: str) -> Tuple[Optional[str], Optional[str]]:
    """Fetches creation and expiration dates from WHOIS data."""
    try:
        whois_data = whois(domain)
        creation_date = str(whois_data.creation_date[0]) if whois_data.creation_date else None  # Handle list or single date
        expiry_date = str(whois_data.expiration_date[0]) if whois_data.expiration_date else None  # Handle list or single date
        return creation_date, expiry_date
    except Exception as e:
        logging.warning(f"Error fetching WHOIS data for {domain}: {e}")
        return None, None

async def update_dataframe_with_whois(df: pd.DataFrame, url_column: str, batch_size: int = 500, wait_time: int = 60) -> pd.DataFrame:
    """Updates the DataFrame with creation_date and expiry_date columns."""

    urls = df[url_column].tolist()
    creation_dates = []
    expiry_dates = []

    for i in range(0, len(urls), batch_size):
        batch = urls[i:i + batch_size]
        logging.info(f"Processing batch {i // batch_size + 1} of {len(urls) // batch_size + 1} - URLs: {i + 1} to {min(i + batch_size, len(urls))}")

        batch_results = await asyncio.gather(*(fetch_whois_data(urlparse(url).netloc) for url in batch))

        creation_dates.extend([result[0] for result in batch_results])
        expiry_dates.extend([result[1] for result in batch_results])

        if i + batch_size < len(urls):
            logging.info(f"Waiting for {wait_time} seconds...")
            await asyncio.sleep(wait_time)

    df['creation_date'] = creation_dates
    df['expiry_date'] = expiry_dates

    return df

```

```python
from urllib.parse import urlparse, parse_qs, urlunparse

def extract_url_features(df: pd.DataFrame, url_column: str) -> pd.DataFrame:
    def ensure_url_scheme(url: str) -> str:
        """Ensure the URL has a scheme, default to 'http' if missing."""
        parsed_url = urlparse(url)
        if not parsed_url.scheme:
            # Add a default scheme (http) if missing
            url = 'http://' + url
        return url

    # Apply the ensure_url_scheme function to ensure all URLs have a scheme
    df[url_column] = df[url_column].apply(ensure_url_scheme)

    # Create feature columns
    df['url_length'] = df[url_column].apply(len)
    df['num_subdirectories'] = df[url_column].apply(lambda x: len(urlparse(x).path.split('/')) - 1 if urlparse(x).path else 0)
    df['num_query_params'] = df[url_column].apply(lambda x: len(parse_qs(urlparse(x).query)))

    # Updated to handle URLs with fewer than 3 parts after splitting
    df['has_ip_address'] = df[url_column].apply(lambda x: any(char.isdigit() for char in x.split('/')[2])
                                                if len(x.split('/')) >= 3 else False)

    df['is_https'] = df[url_column].apply(lambda x: 1 if urlparse(x).scheme == 'https' else 0)
    df['has_www'] = df[url_column].apply(lambda x: 1 if 'www.' in x else 0)
    df['num_subdomains'] = df[url_column].apply(lambda x: len(urlparse(x).netloc.split('.')) - 1 if urlparse(x).netloc else 0)  # Count subdomains, -1 for the domain
    df['domain_length'] = df[url_column].apply(lambda x: len(urlparse(x).netloc.split('.')[0]))  # Length of the domain
    df['tld'] = df[url_column].apply(lambda x: urlparse(x).netloc.split('.')[-1])  # Get TLD
    return df
```

```python
import pandas as pd
import math
from urllib.parse import urlparse
from collections import Counter
import tldextract  # make sure tldextract is installed

def calculate_entropy(domain: str) -> float:
    if not isinstance(domain, str) or not domain:  # Handle non-string or empty domains
        return 0.0
    frequency = Counter(domain)
    length = len(domain)
    entropy = -sum((freq / length) * math.log2(freq / length) for freq in frequency.values())
    return entropy

def extract_domain_features(df: pd.DataFrame, url_column: str) -> pd.DataFrame:
    # Extract domain-related features using tldextract for reliable parsing
    df['domain_name'] = df[url_column].apply(lambda x: tldextract.extract(x).domain)

    df['domain_length'] = df['domain_name'].apply(lambda x: len(str(x)) if x else 0)  # Ensure string and handle null
    df['domain_entropy'] = df['domain_name'].apply(calculate_entropy)
    df['has_hyphen'] = df['domain_name'].apply(lambda x: 1 if isinstance(x, str) and '-' in x else 0)  # Ensure string and handle null
    df['char_repetition'] = df['domain_name'].apply(lambda x: sum(1 for char, count in Counter(str(x)).items() if count > 1) if x else 0)  # Ensure string and handle null

    # Updated phishing words list with commas fixed
    phishing_words = ['bank', 'secure', 'account', 'login', 'label', 'invoice',
                      'post', 'document', 'postal', 'calculations', 'copy',
                      'fedex', 'statement', 'financial', 'dhl', 'usps',
                      'notification', 'irs', 'ups', 'delivery', 'ticket',
                      'urgent', 'important']
    df['common_phishing_words'] = df['domain_name'].apply(lambda x: sum(word in str(x) for word in phishing_words) if x else 0)  # Ensure string and handle null


    return df
```

```python
import aiohttp
import asyncio
import pandas as pd
from urllib.parse import urlparse, urlunparse
from typing import List, Tuple
import logging
from bs4 import BeautifulSoup
import time  # Ensure `time` is imported

# Setup logging with detailed format
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

class ScraperConfig:
    def __init__(self, batch_size: int = 50, retries: int = 3, wait_time: float = 1.0, timeout: int = 30, concurrency: int = 10):
        self.batch_size = batch_size
        self.retries = retries
        self.wait_time = wait_time
        self.timeout = timeout
        self.concurrency = concurrency

def _normalize_url(url: str) -> Tuple[str, str]:
    """Ensure URL has proper scheme with both http and https"""
    parsed_url = urlparse(url)
    if not parsed_url.scheme:
        url_with_http = urlunparse(parsed_url._replace(scheme='http'))
        url_with_https = urlunparse(parsed_url._replace(scheme='https'))
        return url_with_http, url_with_https
    else:
        return urlunparse(parsed_url._replace(scheme="http")), urlunparse(parsed_url._replace(scheme="https"))

async def fetch_links_metadata(session, url: str, retries: int, wait_time: float, timeout: int) -> Tuple[int, int]:
    """Fetch total and external links metadata from a URL with retries and backoff"""
    http_url, https_url = _normalize_url(url)
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    for attempt in range(retries):
        for normalized_url in (http_url, https_url):
            try:
                async with session.get(normalized_url, headers=headers, timeout=timeout) as response:
                    if response.status == 200:
                        html = await response.text()
                        soup = BeautifulSoup(html, 'html.parser')
                        all_links = soup.find_all('a', href=True)
                        external_links = [link for link in all_links if 'http' in link['href']]
                        return len(all_links), len(external_links)
                    else:
                        logging.warning(f"Failed to fetch {url}: Status {response.status}")
            except Exception as e:
                logging.error(f"Error fetching {url} (Attempt {attempt + 1}): {str(e)}")
                await asyncio.sleep(wait_time)  # Backoff wait
    # Return zeros if all attempts fail
    return 0, 0

async def process_urls(urls: List[str], session: aiohttp.ClientSession, config: ScraperConfig) -> List[Tuple[int, int]]:
    """Process a list of URLs with concurrency control"""
    semaphore = asyncio.Semaphore(config.concurrency)  # Concurrency limit

    async def sem_fetch(url: str) -> Tuple[int, int]:
        async with semaphore:
            return await fetch_links_metadata(session, url, config.retries, config.wait_time, config.timeout)

    return await asyncio.gather(*[sem_fetch(url) for url in urls])

async def process_dataframe(df: pd.DataFrame, url_column: str, config: ScraperConfig) -> pd.DataFrame:
    """Process a DataFrame containing URLs and add link metadata columns"""
    start_time = time.time()
    urls = df[url_column].tolist()

    async with aiohttp.ClientSession() as session:
        results = []
        for i, chunk_start in enumerate(range(0, len(urls), config.batch_size)):
            chunk_urls = urls[chunk_start:chunk_start + config.batch_size]
            chunk_results = await process_urls(chunk_urls, session, config)
            results.extend(chunk_results)

            # Logging progress
            processed = len(results)
            elapsed = time.time() - start_time
            rate = processed / elapsed if elapsed > 0 else 0
            logging.info(f"Processed {processed}/{len(urls)} URLs at {rate:.1f} URLs/sec")

    # Add results to DataFrame
    metadata_df = pd.DataFrame(results, columns=['total_links', 'external_links'])
    final_df = pd.concat([df.reset_index(drop=True), metadata_df], axis=1)

    logging.info(f"Total Time: {time.time() - start_time:.1f} seconds")
    return final_df
```

```python
import pandas as pd
from urllib.parse import urlparse, parse_qs
from Levenshtein import distance as levenshtein_distance
from collections import Counter
import re
import math
from nltk.corpus import words
import nltk

# Download NLTK words once
nltk.download('words')
english_words = set(words.words())

# Character set patterns for detecting homograph spoofing
latin_chars = "abcdefghijklmnopqrstuvwxyz"
cyrillic_chars = "абцдефгхийклмнопрстууввшз"  # Cyrillic characters that look similar to Latin ones
homograph_chars = re.compile(f"[{cyrillic_chars}]", re.IGNORECASE)

# Entropy calculation for randomness detection
def calculate_entropy(s: str) -> float:
    frequency = Counter(s)
    length = len(s)
    return -sum((freq / length) * math.log2(freq / length) for freq in frequency.values())

# Function to detect randomness in strings
def is_random_string(s: str) -> bool:
    def entropy(s):
        frequency = Counter(s)
        length = len(s)
        return -sum((freq / length) * math.log2(freq / length) for freq in frequency.values())

    # Check entropy threshold
    if entropy(s) > 4.5:
        return True

    # Check for dictionary words
    words_in_s = re.findall(r'\b\w+\b', s)
    non_dictionary_count = sum(1 for word in words_in_s if word.lower() not in english_words)
    if non_dictionary_count > len(words_in_s) * 0.5:
        return True

    # Check for excessive length in single "word"
    if any(len(word) > 10 for word in words_in_s):
        return True

    # Check for hexadecimal or base64-like pattern
    if re.search(r'[A-Fa-f0-9]{16,}', s) or re.search(r'[A-Za-z0-9+/]{16,}', s):
        return True

    return False

# Main feature extraction function
def extract_url_features(df: pd.DataFrame, url_column: str) -> pd.DataFrame:
    df = df.copy()

    # Parse URL and calculate features
    df['parsed_url'] = df[url_column].apply(urlparse)
    df['path_length'] = df['parsed_url'].apply(lambda x: len(x.path))
    df['query_params'] = df['parsed_url'].apply(lambda x: parse_qs(x.query))
    df['num_query_params'] = df['query_params'].apply(len)
    df['num_slashes'] = df['parsed_url'].apply(lambda x: x.path.count('/'))
    df['has_ip_address'] = df['parsed_url'].apply(lambda x: bool(re.search(r"\b\d{1,3}(\.\d{1,3}){3}\b", x.netloc)))
    df['shortened_url'] = df['parsed_url'].apply(lambda x: any(shortener in x.netloc for shortener in shorteners))

    # Domain-level features
    df['domain'] = df['parsed_url'].apply(lambda x: x.netloc.split('.')[-2] if len(x.netloc.split('.')) > 1 else x.netloc)
    df['domain_length'] = df['domain'].apply(len)
    df['domain_entropy'] = df['domain'].apply(calculate_entropy)
    df['has_hyphen'] = df['domain'].apply(lambda x: '-' in x)
    df['char_repetition'] = df['domain'].apply(lambda x: sum(1 for char, count in Counter(x).items() if count > 1))
    df['common_phishing_words'] = df['domain'].apply(lambda x: sum(word in x for word in phishing_keywords))
    df['typosquatting_distance'] = df['domain'].apply(lambda domain: min((levenshtein_distance(domain, brand) for brand in known_brands), default=0))
    df['contains_homograph_chars'] = df['domain'].apply(lambda x: bool(homograph_chars.search(x)))

    # Path and query-based features
    df['path_suspicious_keywords'] = df['parsed_url'].apply(lambda x: sum(word in x.path for word in phishing_keywords))
    df['query_suspicious_keywords'] = df['query_params'].apply(lambda params: sum(word in key for key in params for word in phishing_keywords))
    df['has_redirect'] = df['parsed_url'].apply(lambda x: 'redirect' in x.path.lower() or 'redirect' in x.query.lower())

    # New features for randomness in URL, title, and description
    df['url_is_random'] = df[url_column].apply(is_random_string)
    df['title_is_random'] = df['title'].apply(is_random_string)
    df['description_is_random'] = df['description'].apply(is_random_string)

    # Clean up unnecessary columns
    df.drop(columns=['parsed_url', 'query_params', 'domain'], inplace=True)

    return df
```
